$\nabla\cdot\left(\sum_{j=0}^Na_j\left(x\right)b_j\left(y\right)\cdot\nabla u\left(x,y\right)\right)=f\left(x,y\right)$


Let us denote $\frac{\partial}{\partial{x}}a\left(x\right)\frac{\partial}{\partial{x}}$ by $\mathcal{D}_a$


$\sum_j^N b_j\left(y\right)\mathcal{D}_{a_j}u\left(x,y\right)+a_j\left(x\right)\mathcal{D}_{b_j}u\left(x,y\right)$


```julia
# a(j,x) = cos.(2pi*j*x)
# b(j,y) = cos.(2pi*j*y)
```

```julia
u(x,y) = sin.(2pi*x).^2.0.*cos.(2pi*y)
```

```julia
lambdify_fast(ex, vars=free_symbols(ex)) = eval(Expr(:function,
                             Expr(:call, gensym(), map(Symbol,vars)...),
                                  convert(Expr, ex)))
```

```julia
# symx = symbols(:symx)
# symy = symbols(:symy)
# symj = symbols(:symj)

# syma = 1.1+cos(2pi*symj*symx)
# symb = 1.1+cos(2pi*symj*symy)
# symu = sin(2pi*symx)^2*cos(2pi*symy)

# symf = diff(syma*symb*diff(symu,symx),symx)+diff(syma*symb*diff(symu,symy),symy)



# # a = lambdify_fast(syma,[symj,symx])
# # b = lambdify_fast(symb,[symj,symy])
# # u = lambdify_fast(symu,[symx,symy])
# f = lambdify_fast(symf,[symj,symx,symy])
```

```julia
symx = symbols(:symx)
symy = symbols(:symy)
symeps1 = symbols(:symeps1)
symeps2 = symbols(:symeps2)
symeps3 = symbols(:symeps3)

syma1 = (1.1+sin(2pi/symeps1*symx))/(1.1+sin(2pi/symeps1*symy))
syma2 = (1.1+sin(2pi/symeps2*symx))/(1.1+cos(2pi/symeps2*symy))
syma3 = (1.1+cos(2pi/symeps3*symx))/(1.1+sin(2pi/symeps3*symy))
symu = sin(2pi*symx)^2*cos(2pi*symy)

symf = diff((syma1+syma2+syma3)*diff(symu,symx),symx)+diff((syma1+syma2+syma3)*diff(symu,symy),symy)
# symf = diff((syma1)*diff(symu,symx),symx)+diff((syma1)*diff(symu,symy),symy)

a1 = lambdify_fast(syma1,[symeps1,symx,symy])
a2 = lambdify_fast(syma2,[symeps2,symx,symy])
a3 = lambdify_fast(syma3,[symeps3,symx,symy])
f = lambdify_fast(symf,[symeps1,symeps2,symeps3,symx,symy])
```

```julia
varx1(ϵ,x) = 1.1.+sin.(2pi/ϵ*x)
vary1(ϵ,y) = 1.0./(1.1.+sin.(2pi/ϵ*y))

varx2(ϵ,x) = 1.1.+sin.(2pi/ϵ*x)
vary2(ϵ,y) = 1.0./(1.1.+cos.(2pi/ϵ*y))

varx3(ϵ,x) = 1.1.+cos.(2pi/ϵ*x)
vary3(ϵ,y) = 1.0./(1.1.+sin.(2pi/ϵ*y))
```

```julia
atol = 1e-5
n = 256
m = 8
hh = 1.0/n
xx = collect(range(0.0,stop=1.0,length=n+1))[1:n]
xxh = collect(range(hh/2,stop=1.0-hh/2,length=n))
yy = collect(range(0.0,stop=1.0,length=n+1))[1:n]
yyh = collect(range(hh/2,stop=1.0-hh/2,length=n))
XX,YY = FactNSF.meshgrid(xx,yy)
;
```

Evaluate functions

```julia
# uu = zeros(n,n)
# aa = zeros(n,n)
# ff = zeros(n,n)
# M = 10

# for i in 1:n
#     for j in 1:n
#         uu[i,j] = u(XX[i,j],YY[i,j])
#         for k in 1:M
#             aa[i,j] += a(k,XX[i,j])*b(k,YY[i,j])
#             ff[i,j] += f(k,XX[i,j],YY[i,j])
#         end
#     end
# end

uu = zeros(n,n)
aa = zeros(n,n)
ff = zeros(n,n)

ϵ1 = 1
ϵ2 = 1
ϵ3 = 1

for i in 1:n
    for j in 1:n
        uu[i,j] = u(XX[i,j],YY[i,j])
        aa[i,j] = a1(ϵ1,XX[i,j],YY[i,j])+a2(ϵ2,XX[i,j],YY[i,j])+a3(ϵ3,XX[i,j],YY[i,j])
        ff[i,j] = f(ϵ1,ϵ2,ϵ3,XX[i,j],YY[i,j])
    end
end
```

```julia
plot(XX,YY,uu,st=:surface)
```

```julia
plot(XX,YY,aa,st=:surface)
```

```julia
plot(XX,YY,ff,st=:surface)
```

```julia
# uvec = [Diagonal(vary1(ϵ1,yy)),FactNSF.get_stagger_grids(6,n,vary1(ϵ1,yyh))]

# vvec = [FactNSF.get_stagger_grids(6,n,varx1(ϵ1,xxh)),Diagonal(varx1(ϵ1,xx))]

# X = FactNSF.TensorDec(n,2,2,[1. for i in 1:2],uvec,vvec)
# FactNSF.normalize(X,atol)

uvec = [Diagonal(vary1(ϵ1,yy)),
    Diagonal(vary2(ϵ2,yy)),
    Diagonal(vary3(ϵ3,yy)),
    FactNSF.get_stagger_grids(6,n,vary1(ϵ1,yyh)),
    FactNSF.get_stagger_grids(6,n,vary2(ϵ2,yyh)),
    FactNSF.get_stagger_grids(6,n,vary3(ϵ3,yyh))]

vvec = [FactNSF.get_stagger_grids(6,n,varx1(ϵ1,xxh)),
    FactNSF.get_stagger_grids(6,n,varx2(ϵ2,xxh)),
    FactNSF.get_stagger_grids(6,n,varx3(ϵ3,xxh)),
    Diagonal(varx1(ϵ1,xx)),
    Diagonal(varx2(ϵ2,xx)),
    Diagonal(varx3(ϵ3,xx))]

X = FactNSF.TensorDec(n,6,6,[1. for i in 1:6],uvec,vvec)
FactNSF.normalize(X,atol)
;
```

# Inverse

```julia
# P = CanTenDec(n,1,1,[1.],
#     [spdiagm(0=>b1(xx))],
#     [spdiagm(0=>a1(yy))])

P = TensorDec(n,3,3,[1.,1.,1.],
    [spdiagm(0=>vary1(ϵ1,yy)),spdiagm(0=>vary2(ϵ2,yy)),spdiagm(0=>vary3(ϵ3,yy))],
    [spdiagm(0=>varx1(ϵ1,xx)),spdiagm(0=>varx2(ϵ2,xx)),spdiagm(0=>varx3(ϵ3,xx))])
FactNSF.normalize(P,atol)

Q = FactNSF.invsqrt(P,1e-2)
;
```

```julia
norm(sum(Q.sig[j]*kron(diag(Q.u[j]),diag(Q.v[j])) for j in 1:Q.rank).^2.0.*sum(P.sig[j]*kron(diag(P.u[j]),diag(P.v[j])) for j in 1:P.rank)-ones(n^2),Inf)
```

```julia
Y = FactNSF.mul(FactNSF.mul(Q,X,atol),Q,atol)
FactNSF.reduce(Y,atol)
println(Y.rank)
```
